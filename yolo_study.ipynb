```python
from ultralytics import YOLO
from google.colab import drive
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/drive')
```

```python
model = YOLO('yolov26n.pt')
```

```python
caminho_imagem = "content/drive/MyDrive/computer-vision-lab/dataset/images/image1.mp4"
```

```python
# roda o modelo no caminho selecionado
resultados = model(caminho_imagem)
```

```python
# plota o resultado da detecção
resultado_anotado = resultados[0].plot()

# converter de BGR para RGB para exibir corretamente com matplotlib
resultado_anotado_rgb = cv2.cvtColor(resultado_anotado, cv2.COLOR_BGR2RGB)

# exibir a imagem anotada
plt.figure(figsize=(10, 10))
plt.imshow(resultado_anotado_rgb)
```
